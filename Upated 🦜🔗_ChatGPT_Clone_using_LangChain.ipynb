```python
from google.colab import files
uploaded = files.upload()

# Get file name
pdf_path = list(uploaded.keys())[0]
print("✅ Uploaded:", pdf_path)
```

```python
from langchain_community.document_loaders import PyPDFLoader

from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Load PDF and split
loader = PyPDFLoader(pdf_path)
pages = loader.load()

splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
texts = splitter.split_documents(pages)

# Embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(texts, embeddings)

# Retriever
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 3})

# LLM - Local Small Model (no API)
llm_pipeline = pipeline("text2text-generation", model="google/flan-t5-small", max_length=512)
llm = HuggingFacePipeline(pipeline=llm_pipeline)

# QA Chain
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)

print("✅ PDF Processed. Ready for questions.")
```

```python
print("\n💡 Ask questions about your PDF. Type 'exit' to quit.\n")

while True:
    question = input("❓ Your Question: ")
    if question.lower() in ['exit', 'quit']:
        print("👋 Bye!")
        break
    try:
        answer = qa_chain.run(question)
        if answer.strip():
            print("🤖", answer)
        else:
            print("❌ ERROR 404")
    except Exception as e:
        print("❌ ERROR 404:", str(e))
```
